## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = '''
SELECT ConvertedComp
FROM master
'''
df = pd.read_sql_query(QUERY,conn)
df.plot(kind='hist')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = '''
SELECT Age
FROM master
'''
df = pd.read_sql_query(QUERY,conn)
df.plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = '''
SELECT Age, WorkWeekHrs
FROM master
'''
df = pd.read_sql_query(QUERY,conn)
# ax1 = df.plot.scatter(x='Age', y='WorkWeekHrs', c='DarkBlue')
plt.scatter('Age', 'WorkWeekHrs', data=df)
plt.xlabel('Age', size=14)
plt.ylabel('Work Week Hours', size=14)
plt.title('Scatter Plot with MatPlotLib', size=18)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = '''
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
'''
df = pd.read_sql_query(QUERY,conn)
df.dropna(inplace=True)
fig = px.scatter(df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', log_x=True, size_max=20)
fig.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = '''
SELECT count(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count(DatabaseDesireNextYear) DESC LIMIT 5
'''

df = pd.read_sql_query(QUERY,conn)
df['Percentage'] = df['Count']/(df['Count'].sum())*100
fig1,ax1 = plt.subplots()
ax1.pie(df['Percentage'], labels=df['DatabaseDesireNextYear'],autopct='%1.1f%%',startangle=90)
ax1.axis('equal')
plt.show()
```

```python
QUERY = '''
SELECT count(LanguageDesireNextYear) as Count, LanguageDesireNextYear
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count(LanguageDesireNextYear) DESC LIMIT 5
'''
df = pd.read_sql_query(QUERY,conn)
df
```

```python
QUERY = '''
SELECT count(LanguageWorkedWith) as Count, LanguageWorkedWith
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY count(LanguageWorkedWith) DESC LIMIT 5
'''
df = pd.read_sql_query(QUERY,conn)
df
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = '''
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age > 29 AND Age < 36
'''
df = pd.read_sql_query(QUERY,conn)
df2 = df.groupby(by='Age').median()
df2.plot(kind='bar', stacked='True')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = '''
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
'''
df = pd.read_sql_query(QUERY,conn)
df2 = df.groupby(['Age']).median()
df2.plot.line()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = '''
SELECT count(MainBranch) as Count, MainBranch
FROM master
'''
df = pd.read_sql_query(QUERY,conn)
plt.barh(df['MainBranch'], df['Count'])
plt.ylabel('Main Title')
plt.xlabel('Count')

plt.show()
```

```python
# your code goes here
QUERY = '''
SELECT COUNT(Respondent) as Count, DevType
FROM DevType
GROUP BY DevType
'''
df = pd.read_sql_query(QUERY,conn)
plt.barh(df['DevType'], df['Count'])
plt.ylabel('Dev Type')
plt.xlabel('Count')

plt.show()
```
